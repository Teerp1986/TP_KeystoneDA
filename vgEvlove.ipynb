```python
Genre= df_vgsales2.groupby('Genre')['Global_Sales'].sum().sort_values(ascending=False)
plt.figure(figsize=(10, 6))
colors = ["#959f99"]*len(Genre)# Create a list of colors for the bars

max_index = Genre.idxmax() 
highlight_position = Genre.index.get_loc(max_index)
colors[highlight_position] = "#17bf39" 

plt.bar(Genre.index, Genre.values, color = colors, width = 0.8) 
ax = plt.gca() # Get the current axes object
ax.spines['top'].set_visible(False) # Remove the top spine
ax.spines['right'].set_visible(False) # Remove the right spine


plt.xlabel("Genre")# Add a label for the x-axis
plt.ylabel("Global Sales (Millions)")# Add a label for the y-axis
plt.title("Highest Grossing Genre in Gaming Globally")# Add a title to the chart
plt.xticks(rotation=45, ha="right")# Rotate x-axis labels for better readability

plt.tight_layout() # Adjust layout to prevent clipping of tick-labels
plt.show()
```

```python

import pandas as pd
import matplotlib.pyplot as plt

df_vgsales2 = pd.read_csv('Data/vgsales2.csv')

# Group, sum, sort, and select top 10
Publisher_sales = (
    df_vgsales2.groupby("Publisher")["Global_Sales"]
    .sum()
    .sort_values(ascending=False)
    .head(10)
    .reset_index()
)

plt.figure(figsize=(10, 6))
plt.plot(
    Publisher_sales["Global_Sales"],
    Publisher_sales["Publisher"],
    marker="o",
    color="#589B4B"
)

plt.title("Top 10 Highest Grossing Game Publishers", fontsize=16)
plt.xlabel("Global Sales (Millions)", fontsize=12)
plt.ylabel("Publisher", fontsize=12)

plt.grid(axis="y", linestyle="--", alpha=0.4)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt

df_vgsales2 = pd.read_csv("Data/vgsales2.csv")

# Create top 10 for each region
global_sales = (
    df_vgsales2.groupby("Platform")["Global_Sales"]
    .sum()
    .sort_values(ascending=False)
    .head(10)
    .reset_index()
)

na_sales = (
    df_vgsales2.groupby("Platform")["NA_Sales"]
    .sum()
    .sort_values(ascending=False)
    .head(10)
    .reset_index()
)

jp_sales = (
    df_vgsales2.groupby("Platform")["JP_Sales"]
    .sum()
    .sort_values(ascending=False)
    .head(10)
    .reset_index()
)

eu_sales = (
    df_vgsales2.groupby("Platform")["EU_Sales"]
    .sum()
    .sort_values(ascending=False)
    .head(10)
    .reset_index()
)

# Create subplots
fig, axes = plt.subplots(2, 2, figsize=(18, 12))

# --- 1. Global: Bar Chart ---
axes[0, 0].bar(global_sales["Platform"], global_sales["Global_Sales"], color="#4C72B0")
axes[0, 0].set_title("Top 10 Platforms – Global Sales (Bar Chart)")
axes[0, 0].tick_params(axis="x", rotation=45)

# --- 2. North America: Scatter Plot ---
axes[0, 1].scatter(na_sales["Platform"], na_sales["NA_Sales"], color="#589B4B", s=120)
axes[0, 1].set_title("Top 10 Platforms – North America (Scatter Plot)")
axes[0, 1].tick_params(axis="x", rotation=45)

# --- 3. Japan: Line Plot ---
axes[1, 0].plot(jp_sales["Platform"], jp_sales["JP_Sales"], marker="o", color="#D95F02")
axes[1, 0].set_title("Top 10 Platforms – Japan (Line Plot)")
axes[1, 0].tick_params(axis="x", rotation=45)

# --- 4. Europe: Horizontal Bar Chart ---
axes[1, 1].barh(eu_sales["Platform"], eu_sales["EU_Sales"], color="#9467BD")
axes[1, 1].set_title("Top 10 Platforms – Europe (Horizontal Bar Chart)")
axes[1, 1].invert_yaxis()  # Highest at top

plt.tight_layout()
plt.show()

```
