```python
# Convert 'Year' column to integer type, filling NaN with 1909. 
# Removed the decimal number in "Year" values.

df_vgsales2 = df_vgsales2.astype({'Year':'float64'})

df_vgsales2['Year'] = df_vgsales2['Year'].fillna('1909')
```

```python
df_vgsales2.head(20)
```

```python
df_vgsales2 = df_vgsales2.astype({'Year':'int64'})

df_vgsales2['Year'] = df_vgsales2['Year'].fillna(1909)
```

```python
df_vgsales2.drop(columns=['Other_Sales'], inplace=True)
```

```python
df_vgsales2.head()
```

```python
df_vgsales2.columns
```

```python
df_vgsales2.rename(columns={'Name':'Game', 'NA_Sales':'North America', 'EU_Sales':'Europe', 'JP_Sales':'Japan'}, inplace=True)
```

```python
df_vgsales2.head(10)
```

```python
# What is the Highest Grossing Genre of Gaming Globally?
sorted_df_vgsales2 = df_vgsales2.sort_values(by=['Genre'], ascending=False) 
Top_10_Genre = sorted_df_vgsales2.head(10)
Top_10_Genre
```

```python
Genre= df_vgsales2.groupby('Genre')['Global_Sales'].sum().sort_values(ascending=False)
plt.figure(figsize=(10, 6))
colors = ["#959f99"]*len(Genre)# Create a list of colors for the bars

max_index = Genre.idxmax() 
highlight_position = Genre.index.get_loc(max_index)
colors[highlight_position] = "#17bf39" 

plt.bar(Genre.index, Genre.values, color = colors, width = 0.8) 
ax = plt.gca() # Get the current axes object
ax.spines['top'].set_visible(False) # Remove the top spine
ax.spines['right'].set_visible(False) # Remove the right spine


plt.xlabel("Genre")# Add a label for the x-axis
plt.ylabel("Global Sales (Millions)")# Add a label for the y-axis
plt.title("Highest Grossing Genre of Gaming Globally")# Add a title to the chart
plt.xticks(rotation=45, ha="right")# Rotate x-axis labels for better readability

plt.tight_layout() # Adjust layout to prevent clipping of tick-labels
plt.show()
```
